# Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = x**2
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')
```

```python
print(x,y)
```

```python
plt.plot(x,y,label="Stock Price 1")
plt.plot(x,x**2 + 3,label="Stock Price 2")
plt.title("My Line Plot")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```

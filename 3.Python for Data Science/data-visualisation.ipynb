```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
# Line Plot
x = np.arange(10)
print(x)
y = x**2
print(y)
z = 2*x + 3
print(z)
```

```python
plt.plot(x,y,color="red",label="Apple",linestyle="dashed",marker="o")
plt.plot(x,z,color="green",label="Kiwi",marker="x")
plt.xlabel("Title")
plt.ylabel("Price")
plt.title("Price of Fruits over Time")
plt.legend()
plt.show()
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
```

```python
plt.plot(prices)
plt.show()
```

## Scatter Plots

```python
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x,y,color="red",label="Apple",linestyle="dashed",marker=">")
plt.scatter(x,z,color="green",label="Kiwi",marker="<")
plt.xlabel("Title")
plt.ylabel("Price")
plt.title("Scatter Plot  -- Price of Fruits over Time")
plt.legend()
plt.show()
```

# Bar Graphs

```python
x_cord = np.array([0,1,2])*2
plt.bar(x_cord,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Silver","Platinum"],color="blue") # Current Year
plt.bar(x_cord+0.5,[20,10,12],width=0.5,label="Next Year",color="orange") # Next Year
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Metal")
plt.legend()
plt.show()
```

# Pie Charts

```python
subjects = "Maths","Chemistry","Physics","Biology"
marks = [70,80,78,90]

plt.pie(marks,labels=subjects,explode=(0,0,0,0.2),autopct="%1.1f%%",shadow=True)
plt.show()
```

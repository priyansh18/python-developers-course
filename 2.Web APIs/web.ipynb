# Web Scrapping

## Beautiful Soup

It is a python library for pulling data out of HTML and XML. It works with your favourite person to provide domestic ways of navigating,searching & modifying the parse tree.It commonly saves programmer hours or fdays of works

```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response = requests.get(url)
```

```python
response.content
```

```python
## HTML Parser
soup = bs4.BeautifulSoup(response.content)
soup
```

```python
div_tags = soup.find("div", {"id": "all_quotes"})
```

```python

inside_divs = div_tags.findAll('div')
```

```python
inside_divs
```

```python
inside_divs[0].img.attrs['src']
```

```python
for i,div in enumerate(inside_divs):
    with open('quotes{}.jpg'.format(i),'wb') as file:
        img_url = div.img.attrs['src']
        response = requests.get(img_url)
        file.write(response.content)
```
